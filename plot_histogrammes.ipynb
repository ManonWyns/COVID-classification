```
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
path = '/content/drive/MyDrive/Projet imagerie /Patients DPI.csv'
df = pd.read_csv(path)

list_DAS = df[['DAS']]
list_DAS = np.array(list_DAS)
plt.figure(1)
plt.hist(list_DAS, range = (0, 3), bins = 4)
plt.xlabel('score')
plt.ylabel('nombre de patients')
plt.title('DAS')

list_DAI = df[['DAI']]
list_DAI = np.array(list_DAI)
plt.figure(2)
plt.hist(list_DAI, range = (0, 3), bins = 4)
plt.xlabel('score')
plt.ylabel('nombre de patients')
plt.title('DAI')

list_DPS = df[['DPS']]
list_DPS = np.array(list_DPS)
plt.figure(3)
plt.hist(list_DPS, range = (0, 3), bins = 4)
plt.xlabel('score')
plt.ylabel('nombre de patients')
plt.title('DPS')

list_DPI = df[['DPI']]
list_DPI = np.array(list_DPI)
plt.figure(4)
plt.hist(list_DPI, range = (0, 3), bins = 4)
plt.xlabel('score')
plt.ylabel('nombre de patients')
plt.title('DPI')

list_GAS = df[['GAS']]
list_GAS = np.array(list_GAS)
plt.figure(5)
plt.hist(list_GAS, range = (0, 3), bins = 4)
plt.xlabel('score')
plt.ylabel('nombre de patients')
plt.title('GAS')

list_GAI = df[['GAI']]
list_GAI = np.array(list_GAI)
plt.figure(6)
plt.hist(list_GAI, range = (0, 3), bins = 4)
plt.xlabel('score')
plt.ylabel('nombre de patients')
plt.title('GAI')

list_GPS = df[['GPS']]
list_GPS = np.array(list_GPS)
plt.figure(7)
plt.hist(list_GPS, range = (0, 3), bins = 4)
plt.xlabel('score')
plt.ylabel('nombre de patients')
plt.title('GPS')

list_GPI = df[['GPI']]
list_GPI = np.array(list_GPI)
plt.figure(8)
plt.hist(list_GPI, range = (0, 3), bins = 4)
plt.xlabel('score')
plt.ylabel('nombre de patients')
plt.title('GPI')

```
