```python
import os, sys
from skimage.io import imread,imsave,imshow
import numpy as np
from skimage.color import rgb2gray
from matplotlib import pyplot as plt
import PIL
import random
from scipy import ndimage
import matplotlib.pyplot as plt
%matplotlib inline

export_path="/Users/manon/DPI/ValidationFlip"
path = "/Users/manon/DPI/Validation DPI 195-255"
dirs = os.listdir(path)
print(len(dirs))
print(dirs[0:10])


print(os.path.join(path,'.DS_Store'))
print(path + '/.DS_Store')

for item in dirs:
    r = random.randint(0,1)
    a = random.randint(0,1)
    if os.path.join(path, item) == path + '/.DS_Store':
        continue
    if os.path.join(path, item) == path + '/Resize.ipynb':
        continue
    if os.path.join(path, item) == path + '/.ipynb_checkpoints':
        continue
    im = imread(os.path.join(path, item), plugin='matplotlib')
    imResize = (im[40:540, 150:650]).astype('uint8')
    gray = rgb2gray(imResize)
    
    if r==0 :
        imsave(os.path.join(export_path, item), gray)

    if r==1 :
        flip = np.fliplr(gray)
        imsave(os.path.join(export_path, item), flip)
       
```
